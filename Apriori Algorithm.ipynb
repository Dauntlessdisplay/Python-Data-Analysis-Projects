```python
from apyori import apriori
```

# 
1. we need to fix the minimum support and the minimum confidence 
2. we will only use the shows with more than 2 users have watched
3. we will re run this model every week which means the recomendation will only remain for 7 days
4. to calculate the minimum support we will use (3*7)/total users
5. we will fix the minimum confidence to 0.2 because we need to be atleast 20% confident
6. we will fix the min lift as 3(note: lift can be 3 or more)

```python
(3*7)/len(df1)
```

```python
# min length should be the number or minimum product that is present in our offer 
# max length should be the maximum product in our offer
# buy 1 get 1 will be 2
# buy 2 get 1 will be 3

rules = apriori(show, min_support=0.0021, min_confidence = 0.2, min_lift =3, min_length =2, max_length = 2)
```

```python
results=list(rules)
results
```

```python
lhs =[]
rhs=[]
support =[]
confidence =[]
lift = []
for i in range(0,len(results)):
    lhs.append(tuple(results[i][0])[0])
    rhs.append(tuple(results[i][0])[1])
    support.append(results[i][1])
    confidence.append(tuple(results[i][2][0])[2])
    lift.append(tuple(results[i][2][0])[3])
```
