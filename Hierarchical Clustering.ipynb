```python
from sklearn.preprocessing import MinMaxScaler

scaling=MinMaxScaler()
```

```python
cars_scaled = scaling.fit_transform(cars_cleaned)
cars_scaled = pd.DataFrame(cars_scaled,columns=cars_cleaned.columns)
```

```python
cars_scaled
```

# Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
```

```python
dendrogram = sch.dendrogram(sch.linkage(cars_scaled,method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
```

```python
clusters = hc.fit_predict(cars_scaled)
```

```python
cars['Cluster'] = clusters
```

```python
cars.head()
```

```python
cars.groupby('Cluster').mean()
```

```python
sns.scatterplot(cars['type'],cars['mpg'],hue=cars['Cluster'])
```

```python
sns.scatterplot(cars['sales'],cars['mpg'],hue=cars['Cluster'])
```
